# ***Librerias***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
```

# ***Carga de Archivo***

```python
url='https://raw.githubusercontent.com/felixhenao/PROYECTO_ANALISIS/refs/heads/main/Reporte_De_Errores_MinTIC_20250421%20ACTU.csv'
df_Errores_Datos = pd.read_csv(url,sep=';')
df_Errores_Datos
```

**Descripción de cada error con su respectivo conteo**

```python
df_Errores_Datos["descripcion"].value_counts()
```

**Validación de valores nulos**

```python
valores_nulos_por_columna = df_Errores_Datos.isnull().sum()
print(valores_nulos_por_columna)
```

**Validación nombres originales de las columnas**

```python
df_Errores_Datos.columns
```

**Corrección nombres de las columnas**

```python
df_Errores_Datos.rename(columns={'c_digo_dane_departamento': 'codigo dane departamento',
                                 'c_digo_dane_municipio':'codigo dane municipio'}
                                 ,inplace=True)
```

**Identificación tipos de datos**

```python
df_Errores_Datos.info()
```

**Conjunto de errores con su respectivo conteo**

```python
top_errores = df_Errores_Datos.groupby('error')['error'].count().reset_index(name='Conteo')
print(top_errores)
```

**Top 10 de errores (Mayor a menor)**

```python
top_errores = top_errores.sort_values('Conteo', ascending=False)
top_errores = top_errores.head(10)
top_errores
```

**Cambio de Bogotá D.C. por Cundinamarca en la columna Departamento**

```python
df_Errores_Datos['departamento'] = df_Errores_Datos['departamento'].replace('Bogotá D.C.', 'Cundinamarca')
df_Errores_Datos['departamento']
```

**Total de errores por Departamento en orden alfabético**

```python
errores_por_departamento = df_Errores_Datos.groupby('departamento')['error'].count().reset_index(name='Conteo')
print(errores_por_departamento)
```

**Top 10 de errores por Departamento (Mayor a menor)**

```python
top_errores_departamento = errores_por_departamento.sort_values('Conteo', ascending=False)
top_errores_departamento = top_errores_departamento.head(10)
top_errores_departamento
```

**Total de errores por Municipio en orden alfabético**

```python
errores_por_municipio = df_Errores_Datos.groupby('municipio')['error'].count().reset_index(name='Conteo')
print(errores_por_municipio)
```

**Top 10 de errores por Municipio (Mayor a menor)**

```python
top_errores_municipio = errores_por_municipio.sort_values('Conteo', ascending=False)
top_errores_municipio = top_errores_municipio.head(10)
top_errores_municipio
```

**Cantidad de errores sobre los 22 existentes por Entidad**

```python
top_errores_por_entidad = df_Errores_Datos.groupby(['entidad', 'error'])['error'].count().reset_index(name='Conteo')
top_errores_por_entidad = top_errores_por_entidad.sort_values('Conteo', ascending=False)
top_10_errores = top_errores_por_entidad.groupby('entidad').size().nlargest(10)
top_10_errores
```

**Total de errores por Entidad**

```python
errores_por_entidad = df_Errores_Datos.groupby('entidad')['error'].count().reset_index(name='Conteo')
print(errores_por_entidad)
```

**Top 10 de errores por Entidad (Mayor a menor)**

```python
top_errores_por_entidad= errores_por_entidad.sort_values('Conteo', ascending=False)
top_errores_por_entidad = top_errores_por_entidad.head(10)
top_errores_por_entidad
```

**Estadisticos descriptivos de la base de datos (errores)**

```python
errores_mas_recurrentes = df_Errores_Datos['error'].value_counts().reset_index(name='Conteo')

# Now you can proceed with sorting and filtering:
errores_mas_recurrentes = errores_mas_recurrentes.sort_values('Conteo', ascending=False)
errores_mas_recurrentes = errores_mas_recurrentes.head(10)
errores_mas_recurrentes
```

**La estandar del top de errores**

**Promedio de los errores mas recurrentes**

```python
promedio_errores_recurrentes = errores_mas_recurrentes['Conteo'].mean().round(2)
print(promedio_errores_recurrentes)
```

**Promedio de errores por departamento**

```python
promedio_errores_por_departamento = errores_por_departamento['Conteo'].mean().round(2)
print(promedio_errores_por_departamento)
```

**Promedio de errores por municipio**

```python
promedio_errores_por_municipio = errores_por_municipio['Conteo'].mean().round(2)
print(promedio_errores_por_municipio)
```

**Varianza del conteo de errores**

**Histograma de errores**

```python
error_counts = df_Errores_Datos['error'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=error_counts.index, y=error_counts.values)
plt.title('Histograma de Errores')
plt.xlabel('Tipo de Error')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Grafico de barras Top 10 errores**

```python
plt.figure(figsize=(12, 6))
top_10_errores.plot(kind='bar')
plt.xlabel("Error")
plt.ylabel("Frecuencia")
plt.title("Top 10 Errors")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Grafico pastel de los 5 errores mas relevantes**

```python
error_counts = df_Errores_Datos['error'].value_counts()
top_5_errors = error_counts.head(5)
labels = top_5_errors.index
sizes = top_5_errors.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Porcentaje de los 5 Errores Más Relevantes')
plt.show()
```

**Grafico de dispersion los 5 errores con mas frecuencia**

```python
error_counts = df_Errores_Datos['error'].value_counts()
top_errors = error_counts.head(5)
top_errors_df = pd.DataFrame({'Error': top_errors.index, 'Frecuencia': top_errors.values})
plt.plot(top_errors_df['Error'], top_errors_df['Frecuencia'], linestyle='--', marker='o', color='green')
plt.title("Los 5 Errores con mas Frecuencia ")
plt.xlabel("Errores")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.annotate('Punto Maximo', xy=(np.pi/2, 1), xytext=(np.pi/-1.5, 1.2), arrowprops=dict(facecolor='red', shrink=0.05))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Distribucion de errores por municipio**

```python

plt.figure(figsize=(8,4))

sns.histplot(data=top_errores_municipio, x='municipio', hue='Conteo', multiple= "stack")

plt.grid()

plt.title("Distribución de errores por Municipio")
plt.xlabel("Municipio")
plt.ylabel("Conteo")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
